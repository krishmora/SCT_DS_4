```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/1122389a-d529-40f7-b166-48110882cc82.csv')

print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nFirst 5 Rows:\n")
print(df.head())

for col in ['Accidents', 'Fatalities', 'Persons Injured']:
    if col in df.columns:
        df[col] = df[col].str.replace(',', '', regex=False).astype(int)

for col in ['% Change from previous year', '% Change from previous year.1', '% Change from previous year.2']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Accidents', data=df, marker='o')
plt.title('Total Accidents Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Fatalities', data=df, marker='o', color='red')
plt.title('Total Fatalities Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Fatalities')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Persons Injured', data=df, marker='o', color='green')
plt.title('Total Persons Injured Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Persons Injured')
plt.grid(True)
plt.show()

print("\n===== ANALYSIS SUMMARY =====")

print("\nTotal Accidents per Year:")
print(df[['Year', 'Accidents']].to_string(index=False))

print("\nTotal Fatalities per Year:")
print(df[['Year', 'Fatalities']].to_string(index=False))

print("\nTotal Persons Injured per Year:")
print(df[['Year', 'Persons Injured']].to_string(index=False))
```
